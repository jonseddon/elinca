## Make background images for the animations

The high resolution background image was the largest image downloaded from https://thomasguymer.co.uk/blog/2018/2018-01-15/ with a URL of https://thomasguymer.co.uk/blog/2018/2018-01-15/natural-earth-1_large.png. It was very hackily added to `envs/elinca_pip/lib/python3.7/site-packages/cartopy/data/raster/natural_earth` and an entry called `large` added to `images.json` in that directory.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
dpi = 100
width_pixels = 800
height_pixels = 800
figure_inches = (width_pixels / dpi, height_pixels / dpi)
fig = plt.figure(figsize=figure_inches, frameon=False)
fig.set_size_inches(width_pixels / dpi, height_pixels / dpi)
ax = fig.add_axes([0., 0., 1., 1.], projection=ccrs.PlateCarree())
# ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([-15, -1, 37, 51], crs=ccrs.PlateCarree())
ax.background_img(resolution='large')
plt.gca().coastlines(resolution='10m')
plt.gca().outline_patch.set_visible(False)
plt.savefig('/home/jseddon/python/elinca/backgrounds/leg_01.png')
```
